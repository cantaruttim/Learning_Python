```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv('AirQualityUCI.csv', sep=';')
dataset
```

```python
dataset.drop(columns=['Unnamed: 15', 'Unnamed: 16'], inplace=True)
```

```python
dataset.dropna(axis=0, inplace=True)
```

```python
dataset.info()
```

```python
dataset
```

```python
# replace ',' por '.'
dataset['CO(GT)'] = dataset['CO(GT)'].str.replace(',', '.')
dataset['C6H6(GT)'] = dataset['C6H6(GT)'].str.replace(',', '.')
dataset['T'] = dataset['T'].str.replace(',', '.')
dataset['RH'] = dataset['RH'].str.replace(',', '.')
dataset['AH'] = dataset['AH'].str.replace(',', '.')

# mudando o formato
dataset['Date'] = pd.Series(dataset['Date'], dtype='datetime64[ns]')
dataset['CO(GT)'] = pd.Series(dataset['CO(GT)'], dtype='float')
dataset['C6H6(GT)'] = pd.Series(dataset['C6H6(GT)'], dtype='float')
dataset['T'] = pd.Series(dataset['T'], dtype='float')
dataset['RH'] = pd.Series(dataset['RH'], dtype='float')
dataset['AH'] = pd.Series(dataset['AH'], dtype='float')
```

```python
dataset.info()
```

```python
dataset3 = dataset[['CO(GT)', 'NMHC(GT)', 'C6H6(GT)', 'NOx(GT)', 'NO2(GT)', 'T', 'RH', 'AH']]
dataset3
```

```python
plt.figure(figsize=(15,7))
sns.pairplot(dataset3, corner=True, diag_kind='kde');
plt.show()
```

```python
dataset2 = dataset[['PT08.S1(CO)', 'PT08.S2(NMHC)', 'PT08.S3(NOx)', 'PT08.S4(NO2)', 'PT08.S5(O3)']]
dataset2
```

```python
plt.figure(figsize=(10,7))
sns.pairplot(dataset2, corner=True, diag_kind='kde');
plt.show()
```
