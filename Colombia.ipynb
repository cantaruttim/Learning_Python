# Base Solicitaciones

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('SOLICITUDES.csv', sep = ";", encoding='utf-8')
dataset
```

```python
# Substituindo o tipo da coluna do dataframe

dataset['Costo unitario'] = dataset['Costo unitario'].str.replace(".","")
dataset['Costo unitario'] = dataset['Costo unitario'].str.replace(",",".")
dataset['Costo unitario'] = dataset['Costo unitario'].astype(float)


dataset['Costo total'] = dataset['Costo total'].str.replace('.','')
dataset['Costo total'] = dataset['Costo total'].str.replace(',','')
dataset['Costo total'] = dataset['Costo total'].astype(float)
```

```python
# Convertendo o tipo da coluna Fecha solicitud para o tipo data
dataset['Fecha solicitud'] = dataset['Fecha solicitud'].astype('datetime64')
```

```python
dataset.to_csv("SOLICITUDES_TRAT.csv")
```

```python
dataset['Cantidad despachada'].describe()
```

```python
df_data = dataset[['Fecha solicitud', 'Cantidad despachada']]
df_data
```

```python
df_data.groupby('Fecha solicitud').count()
```

```python
df_data.info()
```

```python
# Amplitude da base de dados
df_data['Fecha solicitud'].min() , df_data['Fecha solicitud'].max()
```

```python
agrupado = df_data.groupby('Fecha solicitud').sum()
agrupado.to_excel("Datas agrupadas.xlsx")
```

```python
df_data.groupby('Fecha solicitud').sum().plot(title="Quantidade máxima despachada");
plt.xlabel("Date")
plt.ylabel("Cuantidad despachada")
```

```python
qtd_faturada = df_data['Cantidad despachada'].sum()
print(f"A quantidade total faturada de produtos foi de {qtd_faturada} itens")
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(x = df_data['Fecha solicitud'], y = df_data['Cantidad despachada'] );
plt.xlabel("Datas")
plt.ylabel("Quantidades faturadas")
plt.savefig('Dispersão Solicitações Diárias.png', format='png')
```
