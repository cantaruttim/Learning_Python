### Cálculos da Correlação

<br />

- Covariância

$$C(x,y) = \frac{\Sigma(xi - \bar{x}) * (yi - \bar{y}) }{n - 1}$$

<br />

C > 0 : Movimento em conjunto entre as variáveis <br />
C < 0 : Movimento oposto entre as variáveis <br />
C = 0 : As variáveis são independentes

_Obs¹ : As variáveis estão em escalas diferentes, não sendo um bom parâmetro para realizar análises_

<br />

- Coeficiente de Correlação

$$Cr(x,y) = \frac{C(x,y)}{std(x) * std(y)}$$

<br />

_Obs²: Os valores do Coeficiente de Correlação (Cr) estão entre -1 e 1; Quanto mais próximo de 1 as variáveis estão correlacionadas. Quando o valor for igual a 0, as variáveis são independentes. E quando obtemos o valores negativos, uma das variáveis aumenta e a outra diminui_

<br />

- Coeficiente de Determinação

$$Cd(x,y) = Cr²$$

<br />

_Obs³: Indica que a `uma %` da variável dependente consegue ser explicada pelas variáveis explanatórias_ 

<br />

**Correlação NÃO É CAUSA!**

### Regressão

<br />

Objetivo de realizar a previsão de números. Tendo em vista a modelagem de variáveis numéricas (A _variável dependente_ **y** e as _variáveis explanatórias_ **x**)

$$y = b0 + (b1 * x1)$$

<br / > 

Para a regressão linear multipla

$$y = b0 + (b1 * x1) + (b2 * x2) + (b3 * x3) ... (bn * xn)$$


<br />

b0 = _constante_ ; b1 = _coeficiente_

<br />

Para avaliar os residuais uma das técnicas mais utilizadas é o Mean Square Error (MSE): _Penalisa os erros mais altos_

$$RMSE = \frac{1}{N} \sum_{i=1}^N{(fi - yi)²}$$

podemos aplicar tbm o RMSE:

$$MSE = \sqrt{\frac{1}{N} \sum_{i=1}^N{(fi - yi)²}}$$

<br />

Obs: Portanto precisamos encontrar os melhores valores de `b0`e de `b1` para minimizarmos o `MSE`. Com isso temos duas técnicas que podem ser utilizadas. O _Design Matrix_ (Álgebra Linear) que pode ser usada em bases de dados com poucos atributos e trabalha com a inversão de matrizes; _Gradient Descent_ que apresenta um desempenho muito melhor.







```python
import plotly.express as px
import plotly.graph_objects as go
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# análise de regressão simples (com apenas uma variável explanatória - idade)

base_plano_saude = pd.read_csv('./plano_saude.csv')
base_plano_saude
```

```python
X_plano_saude = base_plano_saude.iloc[:, 0].values
X_plano_saude
```

```python
y_plano_saude = base_plano_saude.iloc[:, 1].values
y_plano_saude
```

```python
np.corrcoef(X_plano_saude, y_plano_saude)
```

```python
X_plano_saude.shape
```

```python
X_plano_saude = X_plano_saude.reshape(-1,1) # deixamos a variável X com duas dimensões
X_plano_saude.shape
```

```python
from sklearn.linear_model import LinearRegression
regressor_plano_saude = LinearRegression()
regressor_plano_saude.fit(X_plano_saude, y_plano_saude)
```

```python
# b0
regressor_plano_saude.intercept_
```

```python
# b1
regressor_plano_saude.coef_
```

```python
previsoes = regressor_plano_saude.predict(X_plano_saude)
previsoes
```

```python
X_plano_saude.ravel() # converte para um array 1D
```

```python
grafico = px.scatter(x = X_plano_saude.ravel(), y = y_plano_saude)
grafico.add_scatter(x = X_plano_saude.ravel(), y = previsoes, name = 'Regressão')
grafico.show()
```

```python
regressor_plano_saude.intercept_ + regressor_plano_saude.coef_ * 40
```

```python
regressor_plano_saude.predict([[40]])
```

```python
# indica a qualidade do algorito 0.8 proximo de 1.0
regressor_plano_saude.score(X_plano_saude, y_plano_saude)
```

```python
from yellowbrick.regressor import ResidualsPlot
visualizador = ResidualsPlot(regressor_plano_saude)
visualizador.fit(X_plano_saude, y_plano_saude)
visualizador.poof()
```

### Regressão Polinomial 

$$y = (b0 + b1) * x1 + b2 * x1² ... + bn * x1^{n}$$

```python
base_plano_saude2 = pd.read_csv('plano_saude2.csv')
base_plano_saude2
```

```python
X_plano_saude2 = base_plano_saude2.iloc[:, 0:1].values
y_plano_saude2 = base_plano_saude2.iloc[:, 1].values
```

```python
X_plano_saude2
```

```python
y_plano_saude2
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree = 4)
X_plano_saude2_poly = poly.fit_transform(X_plano_saude2)
```

```python
X_plano_saude2_poly.shape
```

```python
regressor_saude_polinomial = LinearRegression()
regressor_saude_polinomial.fit(X_plano_saude2_poly, y_plano_saude2)
```

```python
# b0
regressor_saude_polinomial.intercept_
```

```python
# b1 (n)
regressor_saude_polinomial.coef_
```

```python
novo = [[40]]

novo = poly.transform(novo)
novo
```

```python
regressor_saude_polinomial.predict(novo)
```

```python
regressor_saude_polinomial.score(X_plano_saude2_poly, y_plano_saude2)
```

```python
previsoes = regressor_saude_polinomial.predict(X_plano_saude2_poly)
previsoes
```

```python
grafico = px.scatter(x = X_plano_saude2[:,0], y = y_plano_saude2)
grafico.add_scatter(x = X_plano_saude2[:,0], y = previsoes, name = 'Regressão')
grafico.show()
```
