# Regression discontinuity design

```python
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
# Avaliando a generosidade com o Indice de Felicidade
 
mod = smf.ols(formula = "Score ~ bs(Generosity, knots=[0.15,0.25,0.35,0.45]) + I(Generosity > 0.15)", data=dataset)
res = mod.fit()
print(res.summary())
```

```python
# Plotting fit

# Getting standard errors
prob_se = res.get_prediction().summary_frame()
prob_se['Generosity'] = dataset['Generosity']
prob_se.sort_values(by='Generosity',inplace=True,ignore_index=True)

low = prob_se[prob_se['Generosity'] <= 0.15].copy()
high = prob_se[prob_se['Generosity'] > 0.15].copy()

# Getting effect for threshold bump
coef = res.summary2().tables[1]
ci = coef.iloc[1,4:6].astype(int).to_list()

fig, ax = plt.subplots()

ax.scatter(dataset['Generosity'], dataset['Score'], c='grey',
           edgecolor='k', alpha=0.15, s=5, zorder=1)

ax.axvline(0.15, linestyle='solid', alpha=1.0, 
           color='k',linewidth=1, zorder=2)

ax.fill_between(low['Generosity'],low['mean_ci_lower'],
                low['mean_ci_upper'],alpha=0.6,
                zorder=3, color='darkblue')

ax.fill_between(high['Generosity'],high['mean_ci_lower'],
                high['mean_ci_upper'],alpha=0.6,
                zorder=3, color='red')

ax.set_xlabel('Predicted Generosity')
ax.set_ylabel('Happiness Score')
ax.set_title(f'RDD Reduced Cost Estimate {ci[0]} to {ci[1]} (95% CI)')

ax.text(0.15, 0.15,'Threshold',rotation=90, size=9,
         ha="center", va="center",
         bbox=dict(boxstyle="round", ec='k',fc='grey'))
```
