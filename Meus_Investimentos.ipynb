## Ações de uma futura carteira

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import datetime
import yfinance as yf
import math
```

```python
ativos = ["ELET3.SA", "VALE3.SA", "VIVT3.SA", "BBSE3.SA", "ITSA4.SA", "BBAS3.SA", "CMIG4.SA",
          "BBDC3.SA", "PETR3.SA", "B3SA3.SA", "TAEE11.SA", "BOVA11.SA"]

acoes = pd.DataFrame()
for ativo in ativos:
  acoes[ativo] = yf.download(ativo, start='2016-01-01')['Close']
```

```python
acoes
```

```python
acoes.dropna(inplace=True)
```

```python
acoes.columns[:]
```

```python
# NORMALIZAÇÃO

# usado para realizar a normalização dos dados para futuras comparações entre várias ações de um portfólio
acoes_df_normalizada = acoes.copy()

for i in acoes_df_normalizada.columns[:]:
  acoes_df_normalizada[i] = acoes_df_normalizada[i] / acoes_df_normalizada[i][0]
```

```python
acoes_df_normalizada
```

```python
import plotly.express as px

figura = px.line(title = 'Histórico do preço das ações')
for i in acoes.columns[1:]:
  figura.add_scatter(y = acoes[i], name = i)
figura.update_layout(hovermode="x")
figura.show()
```

```python
figura = px.line(title = 'Histórico do preço das ações - Normalizada')
for i in acoes_df_normalizada.columns[:]:
  figura.add_scatter(y = acoes_df_normalizada[i], name = i)
figura.update_layout(hovermode="x")
figura.show()
```

```python
taxas_retorno = ( acoes / acoes.shift(1) ) - 1
taxas_retorno
```

```python
taxas_retorno.std() * 246
```

```python
(taxas_retorno.std() * math.sqrt(246)) * 100 # Anualização
```

```python
plt.figure(figsize=(15,5))
sns.heatmap(taxas_retorno.corr(), annot=True);
```

```python
pesos = np.array( [0.18, 0.18, 0.27, 0.18, 0.19])
```

```python
pesos.sum()
```

```python
taxa_carteira_atual = taxas_retorno.drop(columns = ['BOVA11.SA', 'PETR3.SA', "BBDC3.SA",  'VIVT3.SA', 
                                                    'VALE3.SA', 'ELET3.SA', 'BBSE3.SA'])
```

```python
taxa_carteira_atual
```

```python
taxa_carteira_atual.cov() * 246 # covariância anual
```

```python
np.dot(taxa_carteira_atual.cov() * 246, pesos )
```

```python
np.dot(pesos, np.dot(taxa_carteira_atual.cov() * 246, pesos )) # valor da variância
```

```python
volatilidade_portfolio = math.sqrt(np.dot(pesos, np.dot(taxa_carteira_atual.cov() * 246, pesos ))) * 100 # valor do desvio padrão (volatilidade)
volatilidade_portfolio
```

## Análise do Risco da carteira

```python
variancia_carteira = (taxa_carteira_atual.var() * math.sqrt(246)) * pesos
variancia_carteira
```

```python
sub1 = variancia_carteira[0] - variancia_carteira[1] - variancia_carteira[2] - variancia_carteira[3] - variancia_carteira[4]
sub1 
```

```python
risco_nao_sistematico = (np.dot(pesos, np.dot(taxa_carteira_atual.cov() * 246, pesos )) - sub1)
risco_nao_sistematico
```
