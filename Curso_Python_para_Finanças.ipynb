## Taxa de Retorno de Ações

- Taxa de retorno simples

$$Retorno Simples = \frac{preço Final - preço Inicial}{preço Inicial}*100$$

Exemplo, em 2015 o valor da ação era 14,99 reais e 16,07 reais em 2020. Dentro desse período, podemos calcular a taxa de retorno simples. 

$$Retorno Simples = \frac{16,07 - 14,99}{14,99}*100 = 7,20$$

Podemos também considerar o ganho de dividendos. Supondo que ganhamos 0,25 centavos de dividendos teríamos:

$$Retorno Simples = \frac{16,07 + 0,25 - 14,99}{14,99}*100 = 8,87$$


Mas caso, tenhamos 100 ações da Gol. Podemos fazer a seguinte conta, considerando uma taxa de corretagem de 10,00: 

$$Retorno Simples = \frac{1629,50 - 1509}{1509}*100 = 7,98$$

#### Taxa de Retorno Anual

```python
# diária --> anual

dataset['GOLL4.SA']
```

```python
dataset['GOLL4.SA'].shift(1)
```

```python
dataset['RS Gol'] = (dataset['GOLL4.SA'] / dataset['GOLL4.SA'].shift(1)) - 1
dataset['RS CVC'] = (dataset['CVC'] / dataset['CVC'].shift(1)) - 1
dataset['RS WEGE'] = (dataset['WEGE'] / dataset['WEGE'].shift(1)) - 1
dataset['RS MGLU'] = (dataset['MGLU'] / dataset['MGLU'].shift(1)) - 1
dataset['RS TOTVS'] = (dataset['TOTVS'] / dataset['TOTVS'].shift(1)) - 1
dataset['RS BOVA'] = (dataset['BOVA'] / dataset['BOVA'].shift(1)) - 1
```

```python
dataset
```

```python
dataset['RS Gol'].plot(figsize=(15,6));
dataset['RS CVC'].plot(figsize=(15,6));
dataset['RS WEGE'].plot(figsize=(15,6));
dataset['RS MGLU'].plot(figsize=(15,6));
dataset['RS TOTVS'].plot(figsize=(15,6));
dataset['RS BOVA'].plot(figsize=(15,6));
```

Utilizamos a taxa de retorno simples para comparar várias ações e a logarítmica para comparar a mesma ação dentro de um período.

$$RL = log\frac{preçoFinal}{preçoInicial}*100$$
