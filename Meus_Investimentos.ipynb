```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import datetime
import yfinance as yf
```

```python
# Pegando os dados do Yahoo Finance
ativos = ["ELET3.SA", "VALE3.SA", "VIVT3.SA", "BBSE3.SA", "ITSA4.SA", "BBAS3.SA", "CMIG4.SA",
          "BBDC3.SA", "PETR3.SA", "B3SA3.SA", "BOVA11.SA"]

hoje = datetime.datetime.now() # 2023-04-04 01:49:29.384877
quinze_anos_atras = hoje - datetime.timedelta(days = 5475) # 2022-04-04 01:49:29.384877

dados_mercado = yf.download(ativos, quinze_anos_atras, hoje)

display(dados_mercado)
```

```python
dados_fechamento = dados_mercado['Adj Close']
dados_fechamento.columns = ["ELET3.SA", "VALE3.SA", "VIVT3.SA", "BBSE3.SA", "ITSA4.SA", "BBAS3.SA", "CMIG4.SA",
          "BBDC3.SA", "PETR3.SA", "B3SA3.SA", "BOVA11.SA"]

dados_fechamento = dados_fechamento.dropna()
dados_fechamento
```

```python
dados_fechamento.plot(title = "Histórico do Fechamento da minha Carteira", figsize=(17,6));
```

```python
# Criando as tabelas com outros timeframes

dados_fechamento_mensal = dados_fechamento.resample("M").last() # fechamento mensal
dados_fechamento_anual = dados_fechamento.resample("Y").last()

dados_fechamento_anual # fechamento anual
```

```python
# Retorno anual e mensal
retorno_no_ano = dados_fechamento_anual.pct_change().dropna()
retorno_no_mes = dados_fechamento_mensal.pct_change().dropna()
```

```python
ativos = ["ELET3.SA", "VALE3.SA", "VIVT3.SA", "BBSE3.SA", "ITSA4.SA", "BBAS3.SA", "CMIG4.SA",
          "BBDC3.SA", "PETR3.SA", "B3SA3.SA", "BOVA11.SA"]

acoes = pd.DataFrame()
for ativo in ativos:
  acoes[ativo] = yf.download(ativo, start='2020-01-01')['Close']
```

```python
acoes
```

```python
acoes.dropna(inplace=True)
```

```python
acoes.columns[:]
```

```python
# NORMALIZAÇÃO

# usado para realizar a normalização dos dados para futuras comparações entre várias ações de um portfólio
acoes_df_normalizada = acoes.copy()

for i in acoes_df_normalizada.columns[:]:
  acoes_df_normalizada[i] = acoes_df_normalizada[i] / acoes_df_normalizada[i][0]
```

```python
acoes_df_normalizada
```

```python
import plotly.express as px

figura = px.line(title = 'Histórico do preço das ações')
for i in acoes.columns[:]:
  figura.add_scatter(y = acoes[i], name = i)
figura.update_layout(hovermode="x")
figura.show()
```

```python
figura = px.line(title = 'Histórico do preço das ações - Normalizada')
for i in acoes_df_normalizada.columns[:]:
  figura.add_scatter(y = acoes_df_normalizada[i], name = i)
figura.update_layout(hovermode="x")
figura.show()
```
